```python
missing_less_than_5_percent = [var for var in df.columns if df[var].isnull().mean()*100 <5 and df[var].isnull().mean()*100>0]
missing_less_than_5_percent= pd.DataFrame(df[missing_less_than_5_percent])
missing_less_than_5_percent
```

```python
missing_0_percent= [var for var in df.columns if df[var].isnull().mean()==0.0]
missing_0_percent= pd.DataFrame(df[missing_0_percent])
missing_0_percent
```

```python
missing_5_to_10_percent = [var for var in df.columns if df[var].isnull().mean()*100 < 10 and df[var].isnull().mean()*100>5]
missing_5_to_10_percent=df[missing_5_to_10_percent]
missing_5_to_10_percent.info()
```

```python
for i in missing_5_to_10_percent:
    mean1=missing_5_to_10_percent[i].mean()
    print("Before",missing_5_to_10_percent[i].var(),end=' ')
    missing_5_to_10_percent[i].fillna(mean1,inplace=True)
    print("After",missing_5_to_10_percent[i].var())
```

```python
for i in missing_more_than20_percent:
    cleaned_dataframe[i]=missing_more_than20_percent[i]
for i in missing_0_percent:
    cleaned_dataframe[i]=missing_0_percent[i]
for i in missing_less_than_5_percent:
    cleaned_dataframe[i]=missing_less_than_5_percent[i]
for i in missing_5_to_10_percent:
    cleaned_dataframe[i]=missing_5_to_10_percent[i]
```
